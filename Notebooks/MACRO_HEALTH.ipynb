- Section to test FREDAPI data outputs

```python
fred = Fred(api_key=fred_key)
macros = {"UNEMPLOYMENT":"UNRATE",
          "GDP":"GDP",
          "CPI":"CPIAUCSL",
          "PCE":"PCEPI",
          "CBOE VOLATILITY":"VIXCLS"}

tables = {}
for key,val in macros.items():
    df = fred.get_series(
        val,
        units="lin",
        frequency="q")
    df.rename(f"{key}", inplace=True)
    tables[key] = df
all_tables = pd.concat(objs=tables.values(), axis=1, join="inner")
all_tables_change = all_tables.pct_change().round(4)
all_tables_change
```
