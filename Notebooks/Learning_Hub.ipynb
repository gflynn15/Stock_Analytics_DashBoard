```python
import dash
from dash import html, dcc, Input, Output, State
import dash_bootstrap_components as dbc
from dash_bootstrap_templates import load_figure_template

###==========================================Learning Hub Content===================================================###
LEARNING_MODULES = {
    "Company Profile": {
        "icon": "🏢",
        "description": "The 'Identity Card' of a business. Before looking at numbers, you must understand what the company actually does.",
        "metrics": ["Industry", "Sector", "Full-Time Employees", "Business Summary"],
        "why_it_matters": "Different sectors have different 'normal' ranges for ratios. A PE ratio of 20 might be cheap for Tech but expensive for Utilities.",
        "analyst_tip": "Always read the Business Summary to see if the company is diversifying or sticking to a dying legacy business."
    },
    "Valuation": {
        "icon": "🏷️",
        "description": "Is the stock a bargain or a ripoff? Valuation measures the price of the stock relative to the company's fundamentals.",
        "metrics": ["PE Ratio", "Price-to-Book", "Enterprise Value", "PEG Ratio"],
        "why_it_matters": "Buying a great company at a terrible price is a bad investment. Valuation helps you find the right entry point.",
        "analyst_tip": "The PEG Ratio is often better than the PE Ratio because it accounts for growth. A PEG under 1.0 is often considered a 'steal'."
    },
    "Profitability": {
        "icon": "💰",
        "description": "How efficient is the money-making machine? These metrics show how much profit is kept from every dollar of sales.",
        "metrics": ["Profit Margins", "Return on Equity (ROE)", "EBITDA", "Gross Profits"],
        "why_it_matters": "High margins act as a 'moat,' protecting the company from competitors and economic downturns.",
        "analyst_tip": "Look for consistent or expanding ROE. A ROE above 15% usually indicates a company with a strong competitive advantage."
    },
    "Financial Strength": {
        "icon": "🛡️",
        "description": "The 'Safety Net.' Can the company survive a recession or a sudden loss of revenue?",
        "metrics": ["Current Ratio", "Debt-to-Equity", "Total Cash", "Quick Ratio"],
        "why_it_matters": "Companies with too much debt can go bankrupt quickly if interest rates rise or sales slow down.",
        "analyst_tip": "A Current Ratio above 1.0 means the company can pay its short-term bills. Below 1.0 is a red flag."
    },
    "Cash Flow": {
        "icon": "🌊",
        "description": "The lifeblood of a business. Cash Flow is harder to manipulate than accounting earnings (EPS).",
        "metrics": ["Free Cash Flow", "Operating Cash Flow"],
        "why_it_matters": "Earnings can be 'faked' with accounting tricks, but cash hitting the bank account is much harder to hide.",
        "analyst_tip": "Positive Free Cash Flow allows a company to pay dividends, buy back shares, or acquire competitors without taking on debt."
    },
    "Growth & Dividends": {
        "icon": "📈",
        "description": "The Future Rewards. Is the company expanding, and are they sharing the wealth with you?",
        "metrics": ["Revenue Growth", "Earnings Growth", "Dividend Yield", "Payout Ratio"],
        "why_it_matters": "Growth drives stock price appreciation. Dividends provide immediate income while you wait for that growth.",
        "analyst_tip": "Check the Payout Ratio. If it's over 80-90%, the dividend might be at risk of being cut if earnings dip."
    },
    "Share Statistics": {
        "icon": "📊",
        "description": "The Supply and Demand of the stock market. Who else is betting for or against this company?",
        "metrics": ["Short Ratio", "Float Shares", "Insider Ownership"],
        "why_it_matters": "High short interest can lead to a 'short squeeze,' but it also means many smart investors think the stock will fall.",
        "analyst_tip": "High Insider Ownership is a great sign; it means the executives' interests are aligned with yours."
    },
    "Risk Assessment": {
        "icon": "⚠️",
        "description": "The Warning Lights. What external and internal factors could derail the investment?",
        "metrics": ["Beta", "Audit Risk", "Board Risk", "Overall Risk"],
        "why_it_matters": "A high Beta means the stock swings wildly. Risk scores help you see if management is 'risky' behind the scenes.",
        "analyst_tip": "Beta > 1.0 is more volatile than the market. Beta < 1.0 is 'defensive' and usually safer during crashes."
    }
}

###==========================================Dash App Setup=======================================================###
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.CYBORG, "https://cdnjs.cloudflare.com/ajax/libs/animate.css/4.1.1/animate.min.css"])
load_figure_template('cyborg')

###==========================================Custom Components===================================================###
def create_lesson_card(title, data):
    card_id = title.lower().replace(" ", "-").replace("&", "and")
    return dbc.Card([
        dbc.CardBody([
            html.Div([
                html.Span(data['icon'], style={"fontSize": "40px", "marginRight": "15px"}),
                html.H3(title, className="card-title", style={"display": "inline-block", "verticalAlign": "middle"}),
            ], style={"marginBottom": "15px"}),
            html.P(data['description'], className="card-text", style={"fontSize": "18px", "opacity": "0.8"}),
            dbc.Button("Enter Lesson", id=f"btn-{card_id}", color="primary", className="mt-auto", style={"borderRadius": "20px"}),
            
            dbc.Collapse([
                html.Hr(style={"borderColor": "rgba(255,255,255,0.1)"}),
                html.H5("🔑 Key Metrics Included:", style={"color": "#eb4904"}),
                html.Ul([html.Li(m) for m in data['metrics']], style={"columns": "2"}),
                html.H5("❓ Why it matters:", style={"color": "#00d2ff"}),
                html.P(data['why_it_matters']),
                html.Div([
                    html.Strong("👨‍🏫 Teacher's Tip: "),
                    html.Span(data['analyst_tip'])
                ], style={"backgroundColor": "rgba(235, 73, 4, 0.1)", "padding": "15px", "borderRadius": "10px", "borderLeft": "5px solid #eb4904"})
            ], id=f"collapse-{card_id}", is_open=False, className="mt-3")
        ])
    ], style={
        "marginBottom": "20px", 
        "backgroundColor": "rgba(255, 255, 255, 0.05)", 
        "border": "1px solid rgba(255, 255, 255, 0.1)",
        "boxShadow": "0 8px 32px 0 rgba(0, 0, 0, 0.37)",
        "backdropFilter": "blur(4px)",
        "borderRadius": "15px",
        "transition": "transform 0.3s ease-in-out"
    }, className="animate__animated animate__fadeInUp")

###==========================================Layout===============================================================###
app.layout = dbc.Container([
    # --- Header Section ---
    dbc.Row([
        dbc.Col([
            html.H1("🎓 THE LEARNING HUB", style={
                'textAlign': 'center', 
                'fontFamily': 'Inter', 
                'fontWeight': '900',
                'background': '-webkit-linear-gradient(45deg, #fce38a, #f38181)',
                '-webkit-background-clip': 'text',
                '-webkit-text-fill-color': 'transparent',
                'fontSize': '5rem',
                'paddingBottom': '10px',
                'paddingTop': '20px'
            }),
            html.P("Welcome to the Financial Classroom. Click on a module below to master your investment metrics.", 
                   style={"textAlign": "center", "fontSize": "22px", "opacity": "0.9", "color": "white"}),
        ], width=12)
    ], className="learning-hub-header animate__animated animate__fadeInDown")
    
    # --- Classroom Grid ---
    dbc.Row([
        dbc.Col(create_lesson_card(title, data), xs=12, md=6, lg=4) 
        for title, data in LEARNING_MODULES.items()
    ], className="g-4"),

    # --- Footer / Quick Tip ---
    html.Br(),
    html.Div([
        html.H4("📖 Pro-Tip for Students:", style={"color": "#eb4904"}),
        html.P("Investing is not about knowing every number; it's about knowing which numbers matter for the specific company you are reviewing.", 
               style={"fontSize": "18px", "fontStyle": "italic"})
    ], style={"textAlign": "center", "marginTop": "50px", "padding": "30px", "backgroundColor": "rgba(255,255,255,0.02)", "borderRadius": "20px"})

], fluid=True, style={"paddingBottom": "100px"})

###==========================================Callbacks============================================================###
for title in LEARNING_MODULES.keys():
    card_id = title.lower().replace(" ", "-").replace("&", "and")
    
    # Needs to be a factory function to capture card_id properly in the loop
    def create_callback(cid):
        @app.callback(
            [Output(f"collapse-{cid}", "is_open"), Output(f"btn-{cid}", "children")],
            [Input(f"btn-{cid}", "n_clicks")],
            [State(f"collapse-{cid}", "is_open")],
            prevent_initial_call=True
        )
        def toggle_collapse(n, is_open):
            if is_open:
                return False, "Enter Lesson"
            return True, "Close Lesson"
    
    create_callback(card_id)

if __name__ == "__main__":
    app.run(jupyter_mode='external', debug=True)
```
